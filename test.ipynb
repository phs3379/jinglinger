4. **자율주행**

* 자율주행
-  자율주행은 사람에 의해 통제받지 않고 탑재된 인공지능 컴퓨터가 스스로 판단하여 교통(운송)수단을 주행하는 것이다. 현재 자율주행이 가장 발단한 교통수단은 철도이다. 그이유는 기차는 정해진 선로로만 주행하기때문에 돌발 상황이 생길 확률이 적기 때문이다. 실제로 국내에서는 용인 경전철, 의정부 경전철, 인천 2호선등 경전철 노선을 위주로 자동화 되어있다. 
image.png
그에 비해 자율주행 발전이 가장 느린 교통수단은 자동차이다. 대부분 정속 주행을 하며 자동차만 다니는 고속도로 같은 곳은 이미 자율주행이 기술적으로 완성단계에 가까운 상황이지만 보행자와 자전거, 모터사이클 등 방해물의 침입이 빈번해 돌발상황이 자주 일어나는 일반 도로에서는 온갖 예외사항에 대한 처리가 필요하다. 게다가 철도나 항공기, 선박의 운행 환경에 비해서 도로 환경의 복잡성은 너무 높고, 도로 위에 차량 밀집도도 상당히 높기 때문에 별의별 상황이 다 발생하는지라 인간의 개입이 전혀 필요없는 자동화를 하는 것이 기술적으로 매우 어렵다.
